```python
#graph of Active cases in each state
state_cases=df_covid_data.groupby(["State"],as_index=False)["Active Cases"].sum()
state_cases
```

```python
#graph of active cases in each state
state_Acases =df_covid_data.groupby(["State"],as_index=False)["Active Cases"].sum().sort_values(ascending=False,by="Active Cases").head(5)
state_Acases
```

```python
import seaborn as sns
sns.barplot(x="State",y="Active Cases",data=state_Acases)
```

```python
import seaborn as sns
df=df_covid_data.groupby("State")["Active Cases"].sum()
df=df.sort_values(ascending=False).head(5)
sns.barplot(data=df)
```

```python
sns.barplot(x="State",y="Active Cases",data=state_Acases)
```

```python
#acive cases in states(5 states with max cases)
import seaborn as sns
df=df_covid_data.groupby("State")["Active Cases"].sum()
df=df.sort_values(ascending=False).head(5)
sns.barplot(data=df)
```

```python
#graph of state and cases cured(top 10 with min cases cured)
import pandas as pd
import matplotlib.pyplot as plt
df1=df_covid_data.groupby("State")["Cured"].sum()
df1=df1.sort_values(ascending=False).head(10)
df1.plot(kind="bar")
plt.xlabel("State")
plt.ylabel("Cured cases")
plt.title("state vs cured  cases")
plt.show()
```

```python

df_new = df_covid_data.groupby("State")["Cured"].sum()
df_new = df_new.sort_values(ascending=False).head(10)
df_new.plot(kind="bar")
plt.xlabel("Active Cases")
plt.ylabel("State")
plt.title("Active Cases vs State")
plt.show()
```
